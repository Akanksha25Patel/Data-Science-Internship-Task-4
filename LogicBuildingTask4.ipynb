Problem 1: Mobile Recharge Validation System

```python
def validate_recharge(amount, plan):
    valid_plans = [199, 299, 399, 599]

    if amount < 50:
        print(" Recharge amount must be at least ₹50.")
        return False

    if plan not in valid_plans:
        print(" Invalid plan selected.")
        return False

    return True


# Retry using while loop
while True:
    try:
        amount = int(input("Enter recharge amount: ₹"))
        plan = int(input("Enter plan (199, 299, 399, 599): "))

        if validate_recharge(amount, plan):
            print(" Recharge successful!")
            break
        else:
            print(" Please try again.\n")

    except ValueError:
        print("⚠️ Please enter valid numeric values.\n")
```

Problem 2: Inventory Reorder Alert System

```python
def check_reorder(stock_dict):
    for product, stock in stock_dict.items():
        if stock < 15:
            print(f"{product}: Reorder Alert")
        else:
            print(f"{product}: Stock OK")


# Example dictionary
inventory = {
    "Rice": 20,
    "Wheat": 10,
    "Sugar": 5,
    "Oil": 18
}

check_reorder(inventory)
```

Problem 3: Student Result Processing System

```python
def process_result(marks_list):
    total = 0

    # calculate total using loop
    for marks in marks_list:
        total += marks

    average = total / len(marks_list)

    if average >= 50:
        return average, "Pass"
    else:
        return average, "Fail"


# Example usage
marks = []

n = int(input("Enter number of subjects: "))

for i in range(n):
    m = float(input(f"Enter marks for subject {i+1}: "))
    marks.append(m)

avg, result = process_result(marks)

print(f"Average Marks: {avg:.2f}")
print(f"Result: {result}")
```

Problem 4: Cab Fare Estimator with Retry Option


```python
def calculate_fare(distance_km, is_peak):
    base_fare = 50
    per_km_rate = 12

    fare = base_fare + (distance_km * per_km_rate)

    if is_peak:
        fare += fare * 0.25  # 25% extra

    return fare


# Retry using while loop
while True:
    try:
        distance = float(input("Enter distance in km: "))
        peak_input = input("Is it peak hour? (yes/no): ").lower()

        if peak_input not in ["yes", "no"]:
            print(" Please enter 'yes' or 'no'.\n")
            continue

        is_peak = peak_input == "yes"

        total_fare = calculate_fare(distance, is_peak)
        print(f" Estimated Cab Fare: ₹{total_fare:.2f}")
        break

    except ValueError:
        print(" Please enter valid numeric distance.\n")
```

Problem 5: Employee Attendance Eligibility Checker


```python
def check_eligibility(attendance_list):
    present_days = 0

    # count present days using loop
    for status in attendance_list:
        if status.upper() == "P":
            present_days += 1

    total_days = len(attendance_list)
    percentage = (present_days / total_days) * 100

    if percentage >= 75:
        return percentage, "Eligible"
    else:
        return percentage, "Not Eligible"


attendance = []

n = int(input("Enter number of working days: "))

for i in range(n):
    day = input(f"Enter attendance for day {i+1} (P/A): ")
    attendance.append(day)

percent, result = check_eligibility(attendance)

print(f"Attendance Percentage: {percent:.2f}%")
print(f"Status: {result}")
```

Problem 6: Password Strength Checker


```python
def check_password_strength(password):
    has_digit = False
    has_special = False
    special_chars = "@#$"

    # check each character using loop
    for ch in password:
        if ch.isdigit():
            has_digit = True
        if ch in special_chars:
            has_special = True

    if len(password) >= 8 and has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password"


# Example usage
pwd = input("Enter your password: ")
result = check_password_strength(pwd)
print(result)
```
